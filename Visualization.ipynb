### Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read the datasets generated by my_tsp.py

```python
df_Berlin52 = pd.read_csv('tsp_params_fitness_Berlin52.csv')
# df_Berlin52.head()

df_eli101 = pd.read_csv('tsp_params_fitness_eli101.csv')
# df_eli101.head()

df_a280 = pd.read_csv('tsp_params_fitness_a280.csv')
# df_a280.head()
```

### Visualization

#### BERLIN52

```python
g = sns.lmplot(x="co_p", y="fitness", hue="crossover", col="select",
               data=df_Berlin52, x_estimator=np.mean, order=2)\
        .set_axis_labels("Cross Over Rate", "Fitness (min)")\
        .set(xlim=(0, 1), ylim=(7000, 22000))

h = sns.lmplot(x="mu_p", y="fitness", hue="crossover", col="select",
               data=df_Berlin52, x_estimator=np.mean, order=2)\
     .set_axis_labels("Mutation Rate", "Fitness (min)")\
     .set(xlim=(0, 1), ylim=(7000, 22000))

i = sns.lmplot(x="gens", y="fitness", hue="crossover", col="select",
               data=df_Berlin52, x_estimator=np.mean, order=2)\
          .set_axis_labels("Number of Generations", "Fitness (min)")\
          .set(xlim=(100, 1000), ylim=(7000, 22000))

j = sns.lmplot(x="gens", y="fitness", hue="elitism", col="select",
               data=df_Berlin52, x_estimator=np.mean, order=2)\
        .set_axis_labels("Number of Generations", "Fitness (min)")\
        .set(xlim=(100, 1000), ylim=(7000, 22000))
```
